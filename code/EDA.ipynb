This `2400` will throw an exception in `strptime` format because hour is expected to go from 0 to 23.

```python
from ibis import _

flights.filter(flights.DepTime == "2400").count()
```

```python
flights = flights.mutate(
    AdjustedDayofMonth=ibis.case()
    .when(flights.DepTime == "2400", flights.DayofMonth + 1)
    .else_(flights.DayofMonth)
    .end()
)
flights = flights.mutate(DepTime=flights.DepTime.replace("2400", "0000"))
flights
```

```python
flights = flights.mutate(
    LocalDepTimestamp=(
        flights.Year.cast("str")
        + "-"
        + flights.Month.cast("str")
        + "-"
        + flights.DayofMonth.cast("str")
        + " "
        + flights.DepTime
    ).as_timestamp("%Y-%-m-%-d %H%M")
)
flights.LocalDepTimestamp
```

```python
import matplotlib.pyplot as plt
from matplotlib import dates

# pick 2020-8-31 as an example
flights_sample = flights.filter([flights.LocalDepTimestamp > ibis.timestamp("2020-8-31"), flights.LocalDepTimestamp < ibis.timestamp("2020-9-1")])
number_of_flights_by_time = flights_sample.group_by(flights_sample.LocalDepTimestamp).aggregate(number_of_flights=_.count()).to_pandas()

def plot_number_of_flights_by_time(timestamp, number_of_flights):
    fig, ax = plt.subplots(figsize=(10, 10))
    plt.plot(timestamp, number_of_flights, "o")
    plt.gcf().autofmt_xdate()
    plt.xlabel("Time")
    plt.ylabel("Number of flights")
    ax.xaxis.set_major_formatter(dates.DateFormatter("%Y-%m-%d %H:%M"))
    return fig, ax

plot_number_of_flights_by_time(number_of_flights_by_time.LocalDepTimestamp, number_of_flights_by_time.number_of_flights)
```

```python
delay_minutes = flights_sample.DepDelay.to_pandas()
fig, ax = plt.subplots(figsize=(10, 10))
plt.hist(delay_minutes, bins=100)
plt.xlabel("Delayed minutes")
plt.ylabel("Number of flights")
plt.show()
```
